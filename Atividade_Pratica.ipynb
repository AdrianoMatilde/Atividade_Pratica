Exercício 1 – Desenvolver uma classe calculadora que faça operções matemáticas utilizando dois números 
inteiros, sendo os dois últimos números de seu RU. Os dois números inteiros devem ser solicitados para o usuário 
digitar. Se o seu RU for zero, substituí-lo(s) pelo número 5. Sendo as possíveis operações matemáticas: soma(+), 
subtração(-), multiplicação(*), divisão(/), expoente (^), resto(%) e raíz quadrada da soma dos dois números 
( sqrt(Num1 + Num2) ). Além destas funcionalidades, o algoritmo deverá ter um MENU que possíbilite ao 
usuário escolher qual o tipo de operação que se deseja realizar e que possibilite ao usuário a digitar os dois números.
Apresentar todas as operações matemáticas da calculadora funcionando!

```python
# Importa o módulo math, que fornece funções matemáticas adicionais.
import math


class Calculadora:  # Foi criada a classe Calculadora que é uma estrutura que define um tipo de objeto com atributos e comportamentos específicos.
    def __init__(self, num1, num2):
        self.num1 = num1  # Armazena o primeiro número na variável de instância num1
        self.num2 = num2  # Armazena o segundo número na variável de instância num2

    def soma(self):
        return self.num1 + self.num2  # Retorna a soma dos dois números

    def subtracao(self):
        return self.num1 - self.num2  # Retorna a subtração dos dois números

    def multiplicacao(self):
        return self.num1 * self.num2  # Retorna a multiplicação dos dois números

    def divisao(self):
        if self.num2 != 0:  # Verifica se o segundo número não é zero
            return self.num1 / self.num2  # Retorna a divisão dos dois números
        else:
            # Retorna uma mensagem de erro se a divisão por zero for tentada
            return "Divisão por zero não é permitida."

    def expoente(self):
        # Retorna o primeiro número elevado ao segundo número
        return self.num1 ** self.num2

    def resto(self):
        # Retorna o resto da divisão do primeiro número pelo segundo número
        return self.num1 % self.num2

    def raiz_quadrada_soma(self):  # Retorna a raiz quadrada da soma dos dois números
        return math.sqrt(self.num1 + self.num2)


# Imprime a mensagem com o nome "Calculadora" para indicar o início da execução do programa
print("Calculadora")
print("Seja bem-vindo!")

# Imprimi na tela um menu de operações que serão realizadas e o programa só vai ser encerrado após ser escolhido a opção '0'.
menu = """ 
Escolha uma operação:
1. Soma
2. Subtração
3. Multiplicação
4. Divisão
5. Expoente
6. Resto
7. Raiz quadrada da soma dos números
0. Sair

Digite o número da operação desejada: 
"""
opcao = input(
    menu)  # Solicita ao usuário que digite o número correspondente à operação desejada

while opcao != '0':
    num1 = int(input("Digite o primeiro número: "))
    num2 = int(input("Digite o segundo número: "))

# Foi criada uma instância da classe Calculadora com os números digitados pelo usuário
    calculadora = Calculadora(num1, num2)

    if opcao == '1':
        print("Resultado da soma:", calculadora.soma())
    elif opcao == '2':
        print("Resultado da subtração:", calculadora.subtracao())
    elif opcao == '3':
        print("Resultado da multiplicação:", calculadora.multiplicacao())
    elif opcao == '4':
        print("Resultado da divisão:", calculadora.divisao())
    elif opcao == '5':
        print("Resultado do expoente:", calculadora.expoente())
    elif opcao == '6':
        print("Resultado do resto:", calculadora.resto())
    elif opcao == '7':
        print("Resultado da raiz quadrada da soma:",
              calculadora.raiz_quadrada_soma())
    else:
        print("Opção inválida. Por favor, escolha uma opção válida.")

    opcao = input(menu)

# Mensagem exibida quando o usuário escolhe a opção de sair (opção '0')
print("Muito Obrigado! Volte sempre...")
```

Exercício 2 – Dada uma determinada equação linear y = ax + bx – c, sendo que os valores para a, b e c serão os 
três primeiros números de seu RU (a = NUM1, b = NUM2, c = NUM3). Caso, algum número do RU seja 
igual a zero, subistituí-lo(s) pelo número 5. Além disso, será preciso criar um vetor aleatório de tamanho 10, 
onde cada posição do vetor conterá os valores de x para a equação linear. Feito isto, fazer um gráfico para 
mostrar os pontos obtidos pela equação linear dentro do plano cartesiano. Por fim, nomear os eixos Y e X do 
gráfico, colocar cores diferentes para os pontos e colocar legenda. Dica: você vai ter no total, 10 pontos no seu 
plano cartesiano, ou seja, 10 pontos serão ilustrados no gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Solicita os valores de a, b e c ao usuário dos 3 primeiros numeros do RU
a = float(input("Digite o primeiro valor do seu RU: "))
b = float(input("Digite o segundo  valor do seu RU: "))
c = float(input("Digite o terceiro valor do seu RU: "))

# Cria um vetor aleatório de tamanho 10 para os valores de x
x = np.random.rand(10)

# Calcula os valores de y utilizando a equação linear
y = a * x + b * x - c

# Gera uma lista de cores aleatórias
cores = [random.choice(['b', 'g', 'r', 'c', 'm', 'y', 'k']) for _ in range(10)]

# Atribui uma cor diferente para cada ponto no gráfico
plt.scatter(x, y, color=cores, label='Pontos')

# Defini os rótulos (nomes) dos eixos
plt.xlabel('Uninter')
plt.ylabel('Programação')

# Adiciona um título ao gráfico
plt.title('Gráfico de Equação Linear')

# Habilita a grade quadriculada
plt.grid(True)

# Adiciona uma legenda referente aos valores de a, b e c
legenda = f'a={a}, b={b}, c={c}'
plt.legend([legenda])

# Exibi o gráfico
plt.show()
```
